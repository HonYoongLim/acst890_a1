```python
#a.
#import pandas package
import pandas as pd
# First import the csv to local instance, then use pd,read_csv to read csv file
dataset=pd.read_csv('student_results.csv')
dataset.head(10) # display the first 10 lines of the csv file
```

```python
#b.
dataset = dataset.fillna(0) #replace NAs with "0"
dataset.head(10)
```

```python
#c.
import numpy as np
#calculate the proportion of each assessment and display as a array
proportion = np.array([[0.07/70],[0.03/30],[0.03/30],[0.02/20],[0.03/30],[0.02/20],[0.024/30],[0.016/20],[0.07/70],[0.03/30],[0.04/50],[0.02/2],[0.6/100]])
# perform matrix multiplication of the proportion to the assessment marks of each student and multiply by 100 to get the total marks
dataset['Total.marks'] = np.dot(dataset.iloc[:,1:14], proportion)*100 
dataset.head(10)
```

```python
#import seaborn package
import seaborn as sns
#plot the boxplot of total marks by 2 different student category
boxplot = sns.boxplot(x="Student category", data=dataset, y="Total.marks",) 
```

```python
#d.
stat_table = dataset.groupby('Class No',as_index=True).agg({"Total.marks":  ["mean", "var", "std"]})
# groupby() function returns a GroupBy object, but essentially describes how the rows of the original data set has been split.
# functions like mean(), var(), std() can be applied to the GroupBy object to obtain summary statistics for each group as built in Pandas package
# agg() function allows multiple statistics to be calculated per group in one calculation
stat_table.describe
#The .describe() function display statistics for stat_table
```

```python
#e.
Grade=[]

#Use if,elif,else function to assign a grade for each student based on total marks
for score in dataset['Total.marks']:
    if score >= 85:
        Grade.append('HD')
    elif 75 <= score < 85:
        Grade.append('D')
    elif 65 <= score < 75:
        Grade.append('Cr')
    elif 50 <= score < 65:
        Grade.append('P')
    else:
        Grade.append('F')
        
dataset['Grade'] = Grade
dataset.head(10)
```

```python
#f.
#import statsmodels.api package
import statsmodels.api as sm
X = dataset["Assessment 1"] #predictor
y = dataset["Final exam"] #response
X = sm.add_constant(X) #constant
model = sm.OLS(y, X) 
results = model.fit()  #perform simples linear regression and fit the model
print(results.summary())
```

```python
#Comment: The coefficient of Assessment 1 is a positive value, this indicates that both variables have a positive linear relationship.
#         The R square is close to 0 (0.097 in this scenario) which explain only little of the response data around its mean. 
#         Hence, it implies the model fits the data poorly.
          
```

```python
#g.
X = dataset[["Assessment 1", "Assessment 3", "Online quiz 1 ", "Group assignment 2"]] #predictors
y = dataset["Final exam"] #response
X = sm.add_constant(X) #constant
model = sm.OLS(y, X)
results = model.fit() #perform multiple linear regression and fit the model
print(results.summary())
```

```python
#h.
Status = []
#Use if,else function to assign status based on Grade of student
for grade in dataset['Grade']:
    if grade == 'F':
        Status.append('incomplete')
    else:
        Status.append('complete')
        
dataset['Status'] = Status
dataset.head(10)
```
